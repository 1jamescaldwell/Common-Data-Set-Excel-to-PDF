This script loads the downloaded data from qlik into the CDS Excel file template.

```python
import pandas as pd
import numpy as np
from dotenv import load_dotenv
import os

load_dotenv()
qlik_data_folder = os.getenv('qlik_data_folder') 
excel_file_path = os.getenv('excel_file_path') 
excel_file_path_backup = os.getenv('excel_file_path_backup') 

# Load in the qlik data
dfs = {}
for file in os.listdir(qlik_data_folder):
    if file.endswith('.xlsx'):
        file_name = "CDS-" +file.split('.')[0] 
        dfs[file_name] = pd.read_excel(os.path.join(qlik_data_folder, file))
```

```python
from openpyxl import load_workbook

wb = load_workbook(excel_file_path_backup)

for file_name, df in dfs.items():
    if file_name not in wb.sheetnames:
        print(f"Sheet '{file_name}' not found in workbook, skipping.")
        continue

    ws = wb[file_name]
    col_ac = 29  # Column AC = column 29

    # Overwrite data starting at row 2 (skip header in row 1)
    for i, value in enumerate(df.iloc[:, 1], start=2):  # adjust df column as needed
        ws.cell(row=i, column=col_ac, value=value)

wb.save(excel_file_path_backup)
```
